```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
# create function
x = Symbol('x')
f = x**2 + 8*sin(x)

# derivative
df = f.diff(x)

# lambdify both
f = lambdify(x, f)
df = lambdify(x, df)

# evaluate
pts = np.linspace(-5, 8)
plt.plot(pts, f(pts), label='f(x)')
plt.plot(pts, df(pts), label='df/dx')
plt.legend()
plt.show()
```

```python
max_iter = 100
momentum = 0.8
moment = [0]
with_momentum = [False, True]

strategy_steps = []

for use_momentum in with_momentum:
    lr = 0.05
    X0 = 10
    cur_iter = 0
    steps = [X0]
    while True:
        # sum last 10 momentum
        m = sum(moment[-10:])
        # get closer to a local min
        grad = lr * df(X0)
        if use_momentum:
            X1 = X0 + m - grad
        else:
            X1 = X0 - grad
        # save step
        steps.append(X1)
        # save grad
        moment.append(-grad)
        # previous moment are : 0.9*(grad N), 0.9*0.9*(grad N-1), etc.
        moment = [momentum*x for x in moment]
        # calculate step size
        step_size = X1 - X0
        # save new step
        X0 = X1
        if cur_iter > max_iter or abs(step_size) < 0.00001:
            print("Convergence done")
            break
        elif cur_iter%40 == 0:
            lr = lr*0.9
            print("Iteration {} has x value {}".format(cur_iter, X1))
        cur_iter = cur_iter + 1
    strategy_steps.append(steps)
    
print("The local minimum occurs at", X1)
```

```python
fig = plt.figure(num=None, figsize=(8, 10), dpi=80, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(2, 1, 1)
steps = strategy_steps[0]
pts = np.linspace(-5, 8)
# change color to see GD progress
color = [item/255 for item in list(range(len(steps)))]
ax1.plot(pts, f(pts), label='f(x)')
ax1.scatter(steps, [f(x) for x in steps], label='GD steps', c=color)
ax1.scatter(steps[0], f(steps[0]), label='Start GD', c='red', s=50)
ax1.scatter(steps[-1], f(steps[-1]), label='End GD', c='yellow', s=50)
plt.title('Gradient Descent without momentum stays in local minimum')
ax1.legend()

ax2 = fig.add_subplot(2, 1, 2)
steps = strategy_steps[1]
pts = np.linspace(-5, 8)
# change color to see GD progress
color = [item/255 for item in list(range(len(steps)))]
ax2.plot(pts, f(pts), label='f(x)')
ax2.scatter(steps, [f(x) for x in steps], label='GD steps', c=color)
ax2.scatter(steps[0], f(steps[0]), label='Start GD', c='red', s=50)
ax2.scatter(steps[-1], f(steps[-1]), label='End GD', c='yellow', s=50)
plt.title('Gradient descent with momentum can avoid local minimum')
ax2.legend()

plt.show()
fig.savefig('../plots/2D_GD.png')
```
