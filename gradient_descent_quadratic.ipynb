```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
# create function
x = Symbol('x')
f = (x-10)**2

# derivative
df = f.diff(x)

# lambdify both
f = lambdify(x, f)
df = lambdify(x, df)

# evaluate
pts = np.linspace(0, 20)
plt.plot(pts, f(pts))
plt.plot(pts, df(pts))
plt.show()
```
