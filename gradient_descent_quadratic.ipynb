```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
# create function
x = Symbol('x')
f = (x-10)**2

# derivative
df = f.diff(x)

# lambdify both
f = lambdify(x, f)
df = lambdify(x, df)

# evaluate
pts = np.linspace(0, 20)
plt.plot(pts, f(pts), label='f(x)')
plt.plot(pts, df(pts), label='df/dx')
plt.legend()
plt.show()
```

```python
max_iter = 300
lr = 0.01
```

```python
X0 = 30
cur_iter = 0
steps = [X0]

while True:
    # get closer to a local min
    X1 = X0 - lr * df(X0)
    # save step
    steps.append(X1)
    # calculate step size
    step_size = abs(X0 - X1)
    # save new step
    X0 = X1
    if cur_iter > max_iter:
        print("Max iter reached")
        break
    if step_size < 0.00001:
        print("Local min reached")
        break
    elif cur_iter%20 == 0:
        print("Iteration {} has x value {}".format(cur_iter, X1))
    cur_iter = cur_iter + 1
        
    
print("The local minimum occurs at", X1)
```

```python
pts = np.linspace(0, 20)
plt.plot(pts, f(pts), label='f(x)')
plt.plot(pts, df(pts), label='df/dx')
plt.plot(steps, [f(x) for x in steps], label='Gradient descent')
plt.legend()
plt.show()
```
